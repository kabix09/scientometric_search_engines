## Count distibution to settings batch

```python
# Funkcja konwertująca string na Counter
def string_to_counter(distribution_str):
    distribution_dict = ast.literal_eval(distribution_str)
    return collections.Counter(distribution_dict)

# Zastosowanie funkcji `string_to_counter` na kolumnie `distribution`
# jesli to jest zakomentowane to distributions sa stringiem
#distribution_df['distribution'] = distribution_df['distribution'].apply(string_to_counter)

# Funkcja do agregowania Counter
def aggregate_counters(counters):
    result_counter = collections.Counter()
    for counter in counters:
        result_counter += counter
    return str(dict(result_counter))

def aggregate_counters_str(distributions_str):
    result_counter = collections.Counter()
    for distribution_str in distributions_str:
        distribution = string_to_counter(distribution_str) # zamiana string na slownik i utworzenie obiektu counter z niego
        result_counter += distribution # aggregacja (dodanie)
    return str(dict(result_counter))

# Grupowanie po kolumnie `settings` i agregowanie `distribution`
aggregated_series = results_df.groupby('settings')['distribution'].apply(aggregate_counters_str).reset_index()
#aggregated_series.to_csv('../distributions.csv', index=False)

aggregated_series['distribution'] = aggregated_series['distribution'].apply(string_to_counter)
#aggregated_series.to_csv('distributions_2.csv', index=False)
# Wyświetlenie zaktualizowanego DataFrame
display(aggregated_series)
```

# Merge files from two computers

```python
import os
import pandas as pd
from tqdm import tqdm

def merge_results(base_path, max_x=305):
    # Iterate over all directories from 0 to max_x
    for x in tqdm(range(0, max_x + 1), desc="Processing directories"):
        # Define paths for results and results_michal
        results_path = os.path.join(base_path, f'results/{x}/results.csv')
        results_michal_path = os.path.join(base_path, f'results_michal/{x}/results.csv')
        
        # Check if both files exist
        if not os.path.exists(results_path) or not os.path.exists(results_michal_path):
            print(f'Skipping {x} as one of the files is missing')
            continue

        # Load the CSV files
        df_results = pd.read_csv(results_path)
        df_results_michal = pd.read_csv(results_michal_path)

        # Sort df_results_michal by query_id in ascending order
        df_results_michal = df_results_michal.sort_values(by='query_id').reset_index(drop=True)

        # Update query_id in df_results_michal to continue from df_results
        #df_results_michal['query_id'] = df_results_michal['query_id'].apply(lambda qid: qid - 490000 + 490000 + len(df_results))

        # Concatenate the two dataframes
        merged_df = pd.concat([df_results, df_results_michal], ignore_index=True)

        # Save the merged dataframe to a new CSV file
        output_path = os.path.join(base_path, f'merged_results/{x}/results.csv')
        os.makedirs(os.path.dirname(output_path), exist_ok=True)
        merged_df.to_csv(output_path, index=False)
        print(f'Merged file saved at {output_path}')
        
        # Delete the original results_michal CSV file
        os.remove(results_michal_path)
        print(f'Deleted {results_michal_path}')

# Example usage:
base_path = '../data'  # Adjust this path as necessary
merge_results(base_path)
```

```python
import os
import pandas as pd
import ast
from collections import Counter
from tqdm import tqdm

def aggregate_distribution_counts(base_path, max_x=305):
    # Initialize an empty Counter to aggregate counts
    aggregated_counter = Counter()

    # Iterate over all directories from 0 to max_x
    for x in tqdm(range(0, max_x + 1), desc="Processing merged results"):
        # Define path for merged_results
        merged_results_path = os.path.join(base_path, f'merged_results/{x}/merged_results.csv')
        
        # Check if the file exists
        if not os.path.exists(merged_results_path):
            print(f'Skipping {x} as the file is missing')
            continue

        # Load the CSV file
        df = pd.read_csv(merged_results_path)

        # Iterate over the 'distribution' column and update the counter
        for dist_str in df['distribution']:
            # Convert the string representation of the dictionary to a dictionary
            dist_dict = ast.literal_eval(dist_str)
            
            # Update the counter with the current distribution dictionary
            aggregated_counter.update(dist_dict)
    
    # Convert the Counter to a list of tuples sorted by ID (key)
    sorted_counts = sorted(aggregated_counter.items())

    break
    # Print or save the results
    for key, count in sorted_counts:
        print(f'{key}: {count}')
    
    # Optionally, save the results to a CSV file
    output_path = os.path.join(base_path, 'aggregated_distribution_counts.csv')
    with open(output_path, 'w') as f:
        f.write('id,count\n')
        for key, count in sorted_counts:
            f.write(f'{key},{count}\n')

    print(f'Results saved to {output_path}')

# Example usage:
base_path = '../data'  # Adjust this path as necessary
aggregate_distribution_counts(base_path)
```
