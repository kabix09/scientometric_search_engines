# Merge files from two computers

```python
import os
import pandas as pd
from tqdm import tqdm

def merge_results(base_path, max_x=305):
    # Iterate over all directories from 0 to max_x
    for x in tqdm(range(0, max_x + 1), desc="Processing directories"):
        # Define paths for results and results_michal
        results_path = os.path.join(base_path, f'results/{x}/results.csv')
        results_michal_path = os.path.join(base_path, f'results_michal/{x}/results.csv')
        
        # Check if both files exist
        if not os.path.exists(results_path) or not os.path.exists(results_michal_path):
            print(f'Skipping {x} as one of the files is missing')
            continue

        # Load the CSV files
        df_results = pd.read_csv(results_path)
        df_results_michal = pd.read_csv(results_michal_path)

        # Sort df_results_michal by query_id in ascending order
        df_results_michal = df_results_michal.sort_values(by='query_id').reset_index(drop=True)

        # Update query_id in df_results_michal to continue from df_results
        #df_results_michal['query_id'] = df_results_michal['query_id'].apply(lambda qid: qid - 490000 + 490000 + len(df_results))

        # Concatenate the two dataframes
        merged_df = pd.concat([df_results, df_results_michal], ignore_index=True)

        # Save the merged dataframe to a new CSV file
        output_path = os.path.join(base_path, f'merged_results/{x}/results.csv')
        os.makedirs(os.path.dirname(output_path), exist_ok=True)
        merged_df.to_csv(output_path, index=False)
        print(f'Merged file saved at {output_path}')
        
        # Delete the original results_michal CSV file
        os.remove(results_michal_path)
        print(f'Deleted {results_michal_path}')
```

```python
base_path = '../data'
merge_results(base_path)
```

# Distribution aggregator

Count number of citations in logs for each settings series

```python
import os
import pandas as pd
import ast
from collections import Counter
from tqdm import tqdm

def process_and_update_distributions(base_path, max_x=305):
    # Define the global output file path
    global_output_path = os.path.join(base_path, 'distribution_merged.csv')

    # Iterate over all directories from 0 to max_x
    for x in tqdm(range(0, max_x + 1), desc="Processing merged results"):
        # Define path for merged_results
        merged_results_path = os.path.join(base_path, f'merged_results/{x}/results.csv')
        
        # Check if the file exists
        if not os.path.exists(merged_results_path):
            print(f'Skipping {x} as the file is missing')
            continue

        # Load the CSV file
        df = pd.read_csv(merged_results_path)

        # Initialize a Counter for this specific settings_id (x)
        distribution_counter = Counter()

        # Iterate over the 'distribution' column and update the counter
        for dist_str in df['distribution']:
            # Convert the string representation of the dictionary to a dictionary
            dist_dict = ast.literal_eval(dist_str)
            
            # Update the counter with the current distribution dictionary
            distribution_counter.update(dist_dict)

        # Prepare the data for saving
        data = {
            'settings_id': [x],
            'distribution': [str(dict(distribution_counter))]  # Convert Counter to a dictionary and then to string
        }

        # Convert to DataFrame
        output_df = pd.DataFrame(data)

        # Check if the global distribution file exists
        if not os.path.exists(global_output_path):
            # If the file doesn't exist, write the header
            output_df.to_csv(global_output_path, index=False, mode='w')
        else:
            # If the file exists, append without writing the header
            output_df.to_csv(global_output_path, index=False, mode='a', header=False)
```

```python
base_path = '../data'
process_and_update_distributions(base_path)
```
