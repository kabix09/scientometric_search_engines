into citation-based distributions.

The transformation replaces distributions of the form:

- `{paper_id: selection_count}`

with citation-frequency distributions:

- `{n_citations: frequency}`

This representation enables direct analysis of how often papers with a given
citation count are selected under each experimental configuration. The results
are saved as a processed artifact and can be reused in downstream evaluation
and visualization steps.
